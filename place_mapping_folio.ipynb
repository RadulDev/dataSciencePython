```python
import folium as fp
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/main_data.csv")
df
```

```python
plant_dist = fp.Map(location=[11.9709, 75.6208],zoom_start=10)
plant_dist
```

```python
for ind in df.index:
  fp.Marker([df['lat'][ind], df['lon'][ind]],tooltip=df['place'][ind]).add_to(plant_dist)

plant_dist
```

```python
plant_dist.save("/content/drive/MyDrive/map.html")
```
